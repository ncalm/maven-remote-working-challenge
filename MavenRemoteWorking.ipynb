# Maven Analytics - Remote Working Survey Challenge

## About the dataset

This dataset contains the results of two surveys of 1,500 remote workers commissioned by the state of New South Wales (NSW). The first took place in August and September 2020, and aimed to capture the remote work experiences and attitudes at the height of the COVID lockdown. Once health restrictions were reduced and a return to lockdown seemed unlikely, a similar survey was conducted between March and April 2021.

To be eligible, survey respondents had to be employed NSW residents with remote working experience in their current job. This means that both unemployed people and those whose job cannot be done remotely were not taken into account. The resulting sample represents roughly 59% of the NSW workforce.

The responses are stored in 2 csv tables, one for each survey.

Each record represents a single respondent, and contains answers to questions on:

- Their employersâ€™ policies, practices, and attitudes for remote work
- Their own attitude towards working remotely
- The amount of time they spent working remotely, and how they spent it
- The barriers they faced when working remotely, and how they have changed
- The best and worst aspects of working remotely
- Their expectations for working remotely in the future

The dataset can be downloaded here:
https://maven-datasets.s3.amazonaws.com/Remote+Working+Survey/NSW+Remote+Working+Survey.zip

## Matching columns between years

First, we'll use fuzzy matching and a little manual work in Excel to match the variables from one year with the next and give the variables shorter names.

```python
coldiff = pd.Series(df_2020_rws.columns.difference(df_2021_rws.columns))

print('There are {} column header differences between the two files'.format(coldiff.count()))
```

So, we need to first create a mapping between the column headers of each file. We'll do this by taking the column headings and fuzzy matching against the headings from the other file. The match score will serve as a useful guide to whether two headings represent the same question or not. It won't be perfect, but it will be a good start.

We can then output the comparison to Excel for some manual work and assigning variable groups to the questions.

```python
# first, convert the columns (which are Index objects) into dataframes
cols20 = pd.DataFrame(df_2020_rws.columns)
cols20.columns = ['col2020']
cols21 = pd.DataFrame(df_2021_rws.columns)
cols21.columns = ['col2021']

# then cross-join the two lists, so each row in the result has a combination of one column from 2020 and one column from 2021
cols20['one'] = 1
cols21['one'] = 1
df_cols_crossjoin = pd.merge(cols20,cols21,on = 'one').drop('one',axis=1)
df_cols_crossjoin.head()
```

Here we use fuzzywuzzy's fuzz.ratio function to get the best match for each question. Using a lambda function with apply like this is not particularly quick.

```python
# apply the fuzzy matching and add a column returning the score
df_cols_crossjoin['fuzz_ratio'] = df_cols_crossjoin.apply(lambda x: fuzz.ratio(x.col2020,
                                                                               x.col2021),
                                                          axis=1)
df_cols_crossjoin.head()
```

We can write the crossjoin of column matches to a csv file and inspect it with Excel.

```python
df_cols_crossjoin.to_csv(s + 'cols_comparison.csv', encoding = 'cp1252', index = False)
```

The method above is comprehensive in that it compares each header from one file with each header from another. 

However, since we're only interested in the top scoring match per field, we can create a shorter file by just extracting the best match.

First, we will need to work with the columns of each file as lists:

```python
col20 = df_2020_rws.columns.tolist()
col21 = df_2021_rws.columns.tolist()
```

We create a dataframe using the col20 list of columns and append to it a dataframe created from the col21 list of columns.

Note that this can be done with a single line, but to maintain readability and attempt to keep under 80 characters per line (per PEP-8), we should split the work onto several lines.

```python
# create dataframes of the columns from each year
df20 = pd.DataFrame({'col_name' : col20, 'year' : 2020})
df21 = pd.DataFrame({'col_name' : col21, 'year' : 2021})

# append the dataframes together
df_colmatch = pd.concat([df20, df21]).reset_index().drop('index', axis=1)

print(df_colmatch.groupby('year')['col_name'].count())

df_colmatch.head()
```

In order to use fuzzy matching more quickly than the lambda method above, we can use process.extract from fuzzywuzzy. 

This method takes an item, then compares it to every item in a list.

So, in this example, we'll take each column header from one year and compare it with all the column headers from the other year.

To do this, we need the column headers as lists.

We can then put both lists in a list - a list of lists - then apply the same matching process in both directions (just in case).

```python
# create a list of the two lists
cols = [col20,col21]

# then, for each of those lists
for ls in cols:
    
    # we want to match items in 2020 with the list for 2021
    # and match items in 2021 with the list for 2020
    if cols.index(ls) == 0:
        matches = cols[1]
    else:
        matches = cols[0]
    
    # for each item in the current list
    for i in ls:
        
        # where the col_name column is the current item and the year is the current year
        # cols.index(ls) = 0 for 2020, 1 for 2021
        mask = (df_colmatch.col_name == i) & (df_colmatch.year == 2020 + cols.index(ls))
        
        # get these columns from the result produced by process.extract
        flds = ['best_match','fuzz_ratio']
    
        # extract the best match (limit=1) at the 0th index 
        # (in order to get the values rather than a list of values)
        df_colmatch.loc[mask,flds] = process.extract(i, 
                                                     matches,
                                                     limit = 1, 
                                                     scorer = fuzz.ratio)[0]

# the output is one row per question, with the best match of that question with the list of
# questions from the other year
df_colmatch
```

We've created a dataframe containing one row per column from each file. Each row has the best match from the other file listed in a separate column, as well as the match score.

This will be far from perfect, but it will serve as a decent guide when trying to standardize the column names.

Now we write the the matches to a csv so we can edit manually in Excel.

```python
df_colmatch.to_csv(s + 'df_colmatch.csv', encoding = 'cp1252', index = False)
```

## After matching columns between years

To make the creation of various data tables easier, we'll stack the data. As far as storage or long-term data architecture is concerned, this is definitely an anti-pattern, but it's usable as an interim measure while we get organized.

In many of the original columns, there are missing values. This can be problematic later and we definitely want to distinguish between values that were missing in the raw data and values that may have become missing due to some processing error. So, first we fill the NA cells in each of the raw dataframes with a default value. Here, a missing value can either mean 'No answer' (i.e. the respondent was asked, but did not respond), or it can mean 'Not applicable', which is usually when the survey routing has bypassed a question because of a response given to an earlier question. For example, if a respondent is not a manager, they don't need to answer questions about managing remote teams.

```python
# add a year column to each source file's dataframe
df_2020_rws['year'] = 2020
df_2021_rws['year'] = 2021

# provide a default value to cells with no value
df_2020_rws.fillna('No answer / not applicable',inplace=True)
df_2021_rws.fillna('No answer / not applicable',inplace=True)

# "melt" (AKA unpivot) each dataframe, keeping the year and response ID on each row
df_melt20 = pd.melt(df_2020_rws, id_vars = ['year','Response ID'])
df_melt21 = pd.melt(df_2021_rws, id_vars = ['year','Response ID'])

# append the dataframes together
df_melt = pd.concat([df_melt20, df_melt21]).reset_index().drop('index', axis=1)

# take a look at the result
df_melt
```

When we create a single file containing both years, we'll want names which are a bit more easy to work with.

So, we'll grab the standardized column names and variable groups from the Excel file we created earlier.

There are some variables where question answers between years are inconsistently formatted. We might be able to reduce some of the inconsistency by applying some simple text cleanup.

We'll merge the standardized variable names with the melted dataset created above. 

Then apply some simple transformations to clean up some of the inconsistencies

```python
# merge the melted dataset with the standardized column names
df_melt2 = pd.merge(df_melt,
                   df_colmatch_std,
                   left_on = ['variable','year'],
                   right_on = ['col_name','year'],
                   how = 'inner').copy()

# .astype(str) is necessary here otherwise numeric values are returned as NaN after .str
df_melt2['value'] = df_melt2['value'].astype(str)

# trim leading and trailing whitespace
df_melt2['value'] = df_melt2['value'].str.strip()

# remove double-spacing inside the string
df_melt2['value'] = df_melt2['value'].replace('  ',' ')

# replace an inconsistency in the gender variable
df_melt2['value'] = df_melt2['value'].replace('Rather not say',
                                              'I would rather not say')

# one year said "or worse", one year said "or less"
df_melt2['value'] = df_melt2['value'].replace('Iâ€™m 50% less productive when working remotely (or worse)',
                                              'Iâ€™m 50% less productive when working remotely (or less)')
```

Some questions have numeric answers, so we don't want that generic text for no answer in there

```python
flds = ['remote_pay_cut_max']
mask = (df_melt2.var_name_edit.isin(flds) \
        | df_melt2.var_name_edit.str.startswith('onsite_time_spent')) \
        & (df_melt2.value == 'No answer / not applicable')

print(df_melt2.loc[mask].var_name_edit.unique())

df_melt2.loc[mask,'value'] = 0
```

```python
# create a series of True/False to search for rows to update
mask = (df_melt2['value'].str.endswith(' Services')) \
        & (df_melt2['var_name_edit'].isin(['industry',
                                          'industry_detail']))

# update the rows by removing the text from the end of the string
df_melt2.loc[mask, 'value'] = df_melt2.loc[mask].value.str[:-len(' Services')]
```

A similar problem exists in the occupation questions, where one year has a suffix " workers" and the other year doesn't

```python
# now do the same for the word 'workers' in occupation variables
mask = (df_melt2['value'].str.lower().str.endswith(' workers')) \
        & (df_melt2['var_name_edit'].isin(['occupation',
                                          'occupation_detail']))

# update the rows by removing the text from the end of the string
df_melt2.loc[mask, 'value'] = df_melt2.loc[mask].value.str[:-len(' Workers')]
```

There is an inconsistency in the org_size question, where one year uses "I am the only employee" and the other uses "I am a sole trader/owner-operator"

```python
mask = (df_melt2['value'] == 'I am the only employee') \
        & (df_melt2['var_name_edit'] == 'org_size')

# update the rows by removing the text from the end of the string
df_melt2.loc[mask, 'value'] = 'I am a sole trader/owner-operator'
```

## Checking and cleaning respondent answers for "time spent"


"Time spent" questions in 2021 specified that they must add up to 24 hours. We should check for any respondents whose answers don't add to 24.

Let's take a look at the questions which should add to 24 hours first.

```python
# so we can see the whole question text in the variable column
pd.set_option('display.max_colwidth', None)

# get these columns from the melted dataframe
flds = ['year','variable','var_name_edit']

# where the standardized name contains the text time_spent
# and the variable name (the question text) contains the text "24 hours"
mask = (df_melt2.var_name_edit.str.contains('time_spent', case = False))\
        & (df_melt2.variable.str.contains('24 hours', case = False))


# show the question text for each variable
df_melt2[mask][flds].groupby(flds).count()
```

First we'll create a new dataframe called df_time to hold just the data for those questions.

Each group of "remote" or "onsite" must add to 24 hours. 

We'll create a new variable called var_name_group so that it can be used as a groupby variable.

In order for the sum aggregation to work correctly, we need to convert the value to float.

```python
# create a new dataframe containing just the rows for these questions
df_time = df_melt2[mask].copy()

# create a new column containing either the word "remote" or "onsite"
# i.e. the first 6 characters of the var_name_edit column
df_time.loc[:,'var_name_group'] = df_time['var_name_edit'].str[:6]

# NA answers are not relevant for this exercise - set them to zero
df_time.loc[df_time.value == 'No answer / not applicable','value'] = 0

# convert the value column to a float so we can sum the column successfully
df_time['value'] = df_time['value'].astype(float)

# check the new variable
df_time.groupby(['var_name_group'])['value'].count()
```

If we sum the hours by group we can see that there are a few with small deviations from 24

```python
# create a group by object to check the sums of each respondent's values by remote/onsite
df_timechk = df_time.groupby(['var_name_group',
                              'Response ID'])['value'].sum().reset_index()

# create a new dataframe containing just those respondents where the sum is not 24
df_timeadj = df_timechk.loc[df_timechk.value != 24].copy()

# show the problematic rows
df_timeadj
```

But all of them round to 24, so we can assume they were just some small errors in recording the times for each item in the group

```python
# show the rows whose rounded sum is not 24 (there are none)
df_timeadj.loc[(df_timeadj.value.round(0) != 24)]
```

In order to avoid unexpected charts where aggregates don't sum to 100% across these questions, we can apply a small adjustment factor to correct the inconsistencies. We'll apply a factor of 24 / sum(value) to each answer which will scale them up or down appropriately by whatever small amount is needed such that the result becomes 24 across each question group.

```python
# calculate the factor
df_timeadj['factor'] = 24.0 / df_timeadj['value']

# since we're starting to change things, let's make a copy so we can trace back in case of error
df_melt3 = df_melt2.copy()

# update var_group_1 in df_melt3 so that it contains the onsite / remote labels for these questions
mask = (df_melt3.var_name_edit.str.contains('time_spent', case = False))\
        & (df_melt3.variable.str.contains('24 hours', case = False))

df_melt3.loc[mask,'var_group_1'] = df_melt3.loc[mask]['var_name_edit'].str[:6]

# left join the two dataframes using equivalent columns
df_melt4 = pd.merge(df_melt3,
                    df_timeadj,
                    how='left',
                    left_on=['var_group_1','Response ID'],
                    right_on=['var_name_group','Response ID'])

# so we can see the column headers of the new df
df_melt4.head()
```

We can see that the merge has created some new columns and renamed the value column to value_x. 

We want to update value_x by multiplying whatever value is in there by the factor calculated above, or by 1 if there is no factor.

```python
basemask = mask.copy()
maskna = basemask & (df_melt4.value_x == 'No answer / not applicable')

df_melt4.loc[maskna,'value_x'] = 0

df_melt4.loc[mask,'value_x'] = (df_melt4.loc[mask]['value_x'].astype(float) 
                                * df_melt4.loc[mask]['factor'].fillna(1))

# get rid of the unnecessary columns
df_melt4 = df_melt4.drop(['var_name_group','value_y','factor'], axis=1)

# reset the name of the value column
df_melt4.rename(columns = {'value_x':'value'}, inplace = True)


# create a filter to check the results
mask = (df_melt4['Response ID'] == 554) & (df_melt4.var_group_1 == 'remote')

# show the results of the udpate on one respondent
df_melt4.loc[mask]
```

We can see above that respondent 554's values for the remote_time_spent questions have been updated slightly using the factor so that they now sum to 24.

In which case, we can divide their values by 5 to get responses per business day, assuming a 5 day week.

We merge the main melted dataset with these errors using a left join, effectively adding a column with the factor to apply (0.2)

```python
# perform the merge and create a new dataframe
df_melt5 = pd.merge(df_melt4, df_timespenterrors2, on=['year','Response ID','var_name_edit'], how='left')

# show the column headers
df_melt5.head()
```

var_group_1_y is a newly added variable and where it is null, it represents rows that were not part of the time spent errors.

So, we filter the data frame for where that column is not NaN and then set the value to be the multiplication of value and factor. 

This effectively shrinks those values in error to one fifth of their previous magnitude.

```python
# create a filter for just those rows with a factor
mask = (df_melt5.var_group_1_y.isna() == False)

# apply the transformation to the value just on the filtered rows
df_melt5.loc[mask,'value'] = df_melt5.loc[mask]['value'].astype(float) * df_melt5.loc[mask]['factor']

# get rid of the extra columns that we no longer need
df_melt5.drop(['var_group_1_y','value_x','value_y','factor'],axis=1,inplace=True)

# reset the column names to how they were before the merge
df_melt5.rename(columns = {'var_group_1_x':'var_group_1'}, inplace = True)

# show the transformed data
df_melt5.loc[mask].head()
```

## best-worst scaling questions

Some questions are asked several times with different options. These are usually referred to as best-worst scaling, and in survey research typically called MaxDiff. 

It's important to be very careful with this kind of data as we can easily make wrong assumptions about how to use the results.

Typically we have a set of items and each item is compared to every other item in the set. In the case of remote vs. onsite, there are 12 statements. Each statement is compared to every other statement through a series of 6 questions that lists 6 of the items each. The respondent is asked to rate the best and worst statement in each set. 

In this particular design, each statement is presented to the respondent 3 times. Because it is exactly 3 times for each statement for every respondent, the design is balanced and we can simply count the number of times a statement was ranked best and subtract from that the number of times it was ranked worst. This gives us a net score for each statement. 

First, we begin by processing the "Remote vs onsite" question. 

```python
# create a filtered copy of the melt4 dataset
mask = (df_melt5.var_name_edit.str[:16] == 'remote_vs_onsite')

# we'll create a new dataset for the best-worst scaled questions
# and begin by adding the remote_vs_onsite variables
df_bw = df_melt5.loc[mask].copy()

# create a new column that extracts the number from the variable name
# e.g. remote_vs_onsite_4_best - the number is 4
# this number represents the order in which the question was asked
# among other questions with the same wording
df_bw['question_order'] = df_bw['var_name_edit'].str[17:18].astype(int)

df_bw['var_group_2'] = df_bw['var_name_edit'].str[19:]

df_bw = df_bw.drop(['col_name'], axis=1)

# for example, we can see the count of responses at best and worst for each statement
df_bw.groupby(['var_group_1','value','var_group_2'])['Response ID'].count()
```

There are two other questions that use best-worst scaling:

1. Barrier to remote working
2. Barrier while onsite while team is remote

```python
# create a mask of rows containing barrier but not containing direction
mask = (df_melt5.var_name_edit.str.contains('barrier')) \
        & (df_melt5.var_name_edit.str.contains('direction') == False)

# check we have the right var_name_edit values
df_melt5.loc[mask].var_name_edit.unique()
```

Now we append these new variables to the df_bw we created above, essentially putting all best-worst scaled questions in the same dataframe

```python
# (we already have a mask defined above)

# we'll create a new dataset for the best-worst scaled questions
# and begin by adding the remote_vs_onsite variables
df_bw = pd.concat([df_bw, df_melt5.loc[mask]])
df_bw.reset_index()

# extract the number from the variable name into the question_order column
df_bw['question_order'] = df_bw.var_name_edit.str.extract('(\d+)').unstack()[0]

# extract the last word from the variable name (least, most, smallest, biggest, etc)
df_bw['var_group_2'] = df_bw['var_name_edit'].str.rsplit('_',n=1,expand=True)[1]


df_bw = df_bw.drop(['col_name'], axis=1)

# for example, we can see the count of responses at best and worst for each statement
df_bw.groupby(['var_group_1','value','var_group_2'])['Response ID'].count()
```

To make data viz slightly easier, and since we won't be actually showing the best/worst, least/most etc values, we'll standardize them so we can create a simpler net score measure in Tableau

```python
# being careful to read the question before assigning a standard name (the standard name itself doesn't really matter as long as it's indicative of intent)
mapdict = {
    'best': 'good',
    'worst': 'bad',
    'most': 'bad',
    'least': 'good',
    'biggest': 'bad',
    'smallest': 'good'
}

df_bw['var_group_3'] = df_bw.var_group_2.map(mapdict)

df_bw.groupby(['var_group_1','var_group_2','var_group_3'])['Response ID'].count()
```

So, the least impactful remote barriers are good, the most impactful are bad.

The best thing about working remotely is good, the worst is bad.

The biggest barriers to collaborating in a hybrid environment are bad, the smallest are good.

Let's write the data to csv so we can investigate in Excel or Tableau

```python
df_bw.to_csv(s + 'df_bw.csv', index = False)
```
